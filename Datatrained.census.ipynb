### Tuning the model using GridSearch 

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
X=data.loc[:,['major_occupation_code_num','citizenship_num','occupation_code','class_of_worker_num','marital_status_num','race_num','Education_Num','age','capital_gains','capital_losses','dividend_from_Stocks','sex_num','weeks_worked_in_year','wage_per_hour','veterans_benefits','business_or_self_employed']]
y=data['income_level_num']
X_train,X_test,y_train,y_test=train_test_split(X,y)
sample_weights=np.zeros(len(y_train))
sample_weights[y_train==0]=1
sample_weights[y_train==1]=3

params={'max_depth':np.arange(3,9,1)}
clf = GridSearchCV(GradientBoostingClassifier(max_depth=7), params,scoring='f1')
clf.fit(X_train,y_train,sample_weight=sample_weights)
print(clf.best_params_)
print(clf.best_score_)

```

```python
from sklearn.model_selection import GridSearchCV 
params={'n_estimators':np.arange(50,81,10)}
clf = GridSearchCV(GradientBoostingClassifier(max_depth=7), params,scoring='f1')
clf.fit(X_train,y_train,sample_weight=sample_weights)
print(clf.best_params_)
print(clf.best_score_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score


model=GradientBoostingClassifier(max_depth=6,n_estimators=70)
model.fit(X_train,y_train,sample_weight=sample_weights)
y_predict_train=model.predict(X_train)
y_predict_test=model.predict(X_test)
score_train=f1_score(y_train,y_predict_train)
score_test=f1_score(y_test,y_predict_test)
print(score_train)
print(score_test)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score

            
sm=SMOTE()
X_res,y_res=sm.fit_resample(X,y)
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res)

params={'max_depth':np.arange(3,9,1)}
clf = GridSearchCV(GradientBoostingClassifier(), params,scoring='f1')
clf.fit(X_train,y_train)
print(clf.best_params_)
print(clf.best_score_)
```
